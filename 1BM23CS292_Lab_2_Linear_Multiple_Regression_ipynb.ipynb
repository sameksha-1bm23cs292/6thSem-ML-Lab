```python
import pandas as pd

df = pd.read_csv('/content/cannada_per_capita_income.csv')
print("DataFrame loaded successfully. First 5 rows:")
print(df.head())
```

```python
X = df['year'].values.reshape(-1, 1)
y = df['per capita income (US$)'].values

print("Shape of X (features):", X.shape)
print("Shape of y (target):", y.shape)

print("\nMissing values in 'year' column:")
print(df['year'].isnull().sum())

print("\nMissing values in 'per capita income (US$)' column:")
print(df['per capita income (US$)'].isnull().sum())
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X, y)

print("Linear Regression model trained successfully.")
```

```python
import numpy as np

year_2020 = np.array([[2020]])
income_2020 = regressor.predict(year_2020)

print(f"Predicted per capita income for 2020: ${income_2020[0]:,.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Plot historical data
plt.scatter(X, y, color='blue', label='Historical Data')

# Plot regression line
plt.plot(X, regressor.predict(X), color='red', label='Regression Line')

# Plot predicted 2020 income
plt.scatter(year_2020, income_2020, color='green', marker='o', s=100, label='Predicted 2020 Income')

# Add title and labels
plt.title('Canada Per Capita Income Prediction')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')

# Add legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()

print("Plot generated successfully, showing historical data, regression line, and 2020 prediction.")
```

```python
import pandas as pd

df = pd.read_csv('/content/salary.csv')
print("DataFrame 'salary.csv' loaded successfully. First 5 rows:")
print(df.head())
```

```python
X = df['YearsExperience'].values.reshape(-1, 1)
y = df['Salary'].values

print("Shape of X (features):", X.shape)
print("Shape of y (target):", y.shape)

print("\nMissing values in 'YearsExperience' column:")
print(df['YearsExperience'].isnull().sum())

print("\nMissing values in 'Salary' column:")
print(df['Salary'].isnull().sum())
```

```python
X_filled = df['YearsExperience'].fillna(df['YearsExperience'].mean()).values.reshape(-1, 1)

print("Missing values in 'YearsExperience' column after imputation:")
print(pd.DataFrame(X_filled).isnull().sum())
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_filled, y)

print("Linear Regression model trained successfully.")
```

```python
import numpy as np

years_of_experience = np.array([[12]])
predicted_salary = regressor.predict(years_of_experience)

print(f"Predicted salary for 12 years of experience: ${predicted_salary[0]:,.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Plot historical data
plt.scatter(X_filled, y, color='blue', label='Historical Data')

# Plot regression line
plt.plot(X_filled, regressor.predict(X_filled), color='red', label='Regression Line')

# Plot predicted salary for 12 years of experience
plt.scatter(years_of_experience, predicted_salary, color='green', marker='o', s=100, label='Predicted Salary for 12 Years Experience')

# Add title and labels
plt.title('Salary vs. Years of Experience with Prediction')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')

# Add legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()

print("Plot generated successfully, showing historical data, regression line, and 12 years experience prediction.")
```

```python
from word2number import w2n

# Fill NaN values with 0 for 'experience' (addressing FutureWarning)
df_hiring['experience'] = df_hiring['experience'].fillna(0)

# Convert word-based numbers to numerical equivalents
def convert_experience_to_num(exp):
    try:
        return w2n.word_to_num(exp)
    except ValueError:
        return exp # Return as is if already a number or 0

df_hiring['experience'] = df_hiring['experience'].apply(convert_experience_to_num)

# Impute missing values in 'test_score' with its mean
df_hiring['test_score'] = df_hiring['test_score'].fillna(df_hiring['test_score'].mean())

# Define features (X) and target (y)
X = df_hiring[['experience', 'test_score', 'interview_score']].values
y = df_hiring['salary'].values

print("Data preparation complete.")
print("Shape of X (features):", X.shape)
print("Shape of y (target):", y.shape)

print("\nMissing values in X after imputation and conversion:")
print(pd.DataFrame(X).isnull().sum())

print("\nMissing values in y:")
print(pd.Series(y).isnull().sum())
```

```python
from sklearn.linear_model import LinearRegression

regressor_multi = LinearRegression()
regressor_multi.fit(X, y)

print("Multiple Linear Regression model trained successfully.")
```

```python
import numpy as np

# Candidate 1: 2 years experience, 9 test score, 6 interview score
candidate1_features = np.array([[2, 9, 6]])
predicted_salary_candidate1 = regressor_multi.predict(candidate1_features)

# Candidate 2: 12 years experience, 10 test score, 10 interview score
candidate2_features = np.array([[12, 10, 10]])
predicted_salary_candidate2 = regressor_multi.predict(candidate2_features)

print(f"Predicted salary for Candidate 1 (2 yrs exp, 9 test, 6 interview): ${predicted_salary_candidate1[0]:,.2f}")
print(f"Predicted salary for Candidate 2 (12 yrs exp, 10 test, 10 interview): ${predicted_salary_candidate2[0]:,.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Calculate predicted salaries for the training data
y_pred_train = regressor_multi.predict(X)

plt.figure(figsize=(10, 6))

# Plot actual vs. predicted salaries for training data
plt.scatter(y, y_pred_train, color='blue', label='Training Data (Actual vs. Predicted)')

# Draw ideal regression line (where actual = predicted)
min_val = np.min([y, y_pred_train])
max_val = np.max([y, y_pred_train])
plt.plot([min_val, max_val], [min_val, max_val], color='black', linestyle='--', label='Ideal Prediction')

# Plot predicted salary for Candidate 1
plt.scatter(predicted_salary_candidate1[0], predicted_salary_candidate1[0], color='green', marker='o', s=100, label='Predicted Salary C1')

# Plot predicted salary for Candidate 2
plt.scatter(predicted_salary_candidate2[0], predicted_salary_candidate2[0], color='red', marker='o', s=100, label='Predicted Salary C2')

# Add title and labels
plt.title('Multiple Linear Regression: Actual vs. Predicted Salaries')
plt.xlabel('Actual Salary ($)')
plt.ylabel('Predicted Salary ($)')

# Add legend and grid
plt.legend()
plt.grid(True)

# Display the plot
plt.show()

print("Plot generated successfully, showing actual vs. predicted salaries and new candidate predictions.")
```
