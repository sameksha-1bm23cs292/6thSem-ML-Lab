```python
import pandas as pd

# Download the 'diabetes.csv' file
!wget -q https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv

# Load the 'diabetes.csv' file into a pandas DataFrame
df_diabetes = pd.read_csv('diabetes.csv')

# Display the first 5 rows of df_diabetes
print("First 5 rows of the DataFrame:")
print(df_diabetes.head())

# Display the summary statistics of df_diabetes
print("\nSummary statistics of the DataFrame:")
print(df_diabetes.describe())

# Display information about the columns, non-null values, and data types of df_diabetes
print("\nInformation about the DataFrame columns and data types:")
df_diabetes.info()
```

```python

import numpy as np

# Columns where 0 values should be considered missing
columns_with_zeros_as_nan = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

# Replace 0s with NaN in the specified columns
df_diabetes[columns_with_zeros_as_nan] = df_diabetes[columns_with_zeros_as_nan].replace(0, np.nan)

print("First 5 rows of df_diabetes after replacing 0s with NaN:")
print(df_diabetes.head())

print("\nNumber of NaN values per column after replacement:")
print(df_diabetes[columns_with_zeros_as_nan].isnull().sum())
```

```python
df_diabetes_cleaned = df_diabetes.copy()

for column in columns_with_zeros_as_nan:
    median_val = df_diabetes_cleaned[column].median()
    df_diabetes_cleaned[column].fillna(median_val, inplace=True)

print("First 5 rows of df_diabetes_cleaned after NaN imputation:")
print(df_diabetes_cleaned.head())

print("\nNumber of NaN values per column after imputation:")
print(df_diabetes_cleaned[columns_with_zeros_as_nan].isnull().sum())
```

```python
df_diabetes_cleaned = df_diabetes.copy()

for column in columns_with_zeros_as_nan:
    median_val = df_diabetes_cleaned[column].median()
    df_diabetes_cleaned[column] = df_diabetes_cleaned[column].fillna(median_val)

print("First 5 rows of df_diabetes_cleaned after NaN imputation:")
print(df_diabetes_cleaned.head())

print("\nNumber of NaN values per column after imputation:")
print(df_diabetes_cleaned[columns_with_zeros_as_nan].isnull().sum())
```

```python
df_diabetes_cleaned_no_outliers = df_diabetes_cleaned.copy()

numerical_cols = df_diabetes_cleaned_no_outliers.columns.drop('Outcome')

for column in numerical_cols:
    Q1 = df_diabetes_cleaned_no_outliers[column].quantile(0.25)
    Q3 = df_diabetes_cleaned_no_outliers[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap the outliers
    df_diabetes_cleaned_no_outliers[column] = df_diabetes_cleaned_no_outliers[column].clip(lower=lower_bound, upper=upper_bound)

print("First 5 rows of df_diabetes_cleaned after outlier capping:")
print(df_diabetes_cleaned_no_outliers.head())

print("\nSummary statistics of df_diabetes_cleaned after outlier capping:")
print(df_diabetes_cleaned_no_outliers.describe())
```

```python
from sklearn.preprocessing import MinMaxScaler

df_diabetes_minmax_scaled = df_diabetes_cleaned_no_outliers.copy()

# Identify numerical columns for scaling (all except 'Outcome')
numerical_cols_for_scaling = df_diabetes_minmax_scaled.columns.drop('Outcome')

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Apply Min-Max scaling to the numerical columns
df_diabetes_minmax_scaled[numerical_cols_for_scaling] = scaler.fit_transform(df_diabetes_minmax_scaled[numerical_cols_for_scaling])

print("First 5 rows of df_diabetes after Min-Max scaling:")
print(df_diabetes_minmax_scaled.head())

print("\nSummary statistics of df_diabetes after Min-Max scaling:")
print(df_diabetes_minmax_scaled.describe())
```

```python
from sklearn.preprocessing import StandardScaler

df_diabetes_standard_scaled = df_diabetes_cleaned_no_outliers.copy()

# Identify numerical columns for scaling (all except 'Outcome')
numerical_cols_for_scaling = df_diabetes_standard_scaled.columns.drop('Outcome')

# Initialize StandardScaler
scaler = StandardScaler()

# Apply Standard scaling to the numerical columns
df_diabetes_standard_scaled[numerical_cols_for_scaling] = scaler.fit_transform(df_diabetes_standard_scaled[numerical_cols_for_scaling])

print("First 5 rows of df_diabetes after Standard scaling:")
print(df_diabetes_standard_scaled.head())

print("\nSummary statistics of df_diabetes after Standard scaling:")
print(df_diabetes_standard_scaled.describe())
```

## Preprocess Adult Income Data

### Subtask:
Implement data cleaning techniques for the Adult Income dataset, including identifying and handling implicit missing values, encoding categorical features, and addressing outliers. Subsequently, apply both Min-Max scaling and Standard scaling to the numerical features.


**Reasoning**:
The previous code block successfully loaded the `df_adult` dataset. Now, I need to identify and handle implicit missing values (like '?' and leading/trailing spaces) in the categorical columns. This will involve cleaning these values and then replacing them with `np.nan`.



```python
import numpy as np

# Identify categorical columns
categorical_cols = df_adult.select_dtypes(include='object').columns

# Clean categorical columns by stripping whitespace
for col in categorical_cols:
    df_adult[col] = df_adult[col].str.strip()

# Replace '?' with NaN in all columns (both categorical and numerical where '?' might appear)
df_adult = df_adult.replace('?', np.nan)

print("First 5 rows of df_adult after cleaning and replacing '?' with NaN:")
print(df_adult.head())

print("\nNumber of NaN values per column after replacement:")
print(df_adult.isnull().sum())
```

**Reasoning**:
Now that implicit missing values have been identified and replaced with `NaN`, I will impute these missing values. I will create a copy of the DataFrame, `df_adult_cleaned`, and then iterate through the categorical columns that contain `NaN` values, filling them with the mode of their respective columns. Since no numerical columns have `NaN` values after the '?' replacement, no median imputation is needed for numerical columns at this stage.



```python
df_adult_cleaned = df_adult.copy()

# Identify columns with NaN values
missing_cols = df_adult_cleaned.columns[df_adult_cleaned.isnull().any()].tolist()

# Impute categorical columns with their mode
for col in missing_cols:
    if df_adult_cleaned[col].dtype == 'object':
        mode_val = df_adult_cleaned[col].mode()[0]
        df_adult_cleaned[col].fillna(mode_val, inplace=True)

print("First 5 rows of df_adult_cleaned after NaN imputation:")
print(df_adult_cleaned.head())

print("\nNumber of NaN values per column after imputation:")
print(df_adult_cleaned.isnull().sum())
```

**Reasoning**:
The previous code block produced a `FutureWarning` because `inplace=True` was used on a sliced DataFrame. To address this, I will modify the line to explicitly assign the result of `fillna` back to the column, thus avoiding the warning and ensuring correct imputation.



```python
df_adult_cleaned = df_adult.copy()

# Identify columns with NaN values
missing_cols = df_adult_cleaned.columns[df_adult_cleaned.isnull().any()].tolist()

# Impute categorical columns with their mode
for col in missing_cols:
    if df_adult_cleaned[col].dtype == 'object':
        mode_val = df_adult_cleaned[col].mode()[0]
        df_adult_cleaned[col] = df_adult_cleaned[col].fillna(mode_val)

print("First 5 rows of df_adult_cleaned after NaN imputation:")
print(df_adult_cleaned.head())

print("\nNumber of NaN values per column after imputation:")
print(df_adult_cleaned.isnull().sum())
```

**Reasoning**:
Now that missing values have been imputed, I will address outliers in the numerical columns of `df_adult_cleaned` by capping them using the Interquartile Range (IQR) method. This will ensure that extreme values do not disproportionately affect subsequent analyses.



```python
df_adult_cleaned_no_outliers = df_adult_cleaned.copy()

# Identify numerical columns for outlier capping
numerical_cols_adult = df_adult_cleaned_no_outliers.select_dtypes(include=['int64', 'float64']).columns.tolist()

for column in numerical_cols_adult:
    Q1 = df_adult_cleaned_no_outliers[column].quantile(0.25)
    Q3 = df_adult_cleaned_no_outliers[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap the outliers
    df_adult_cleaned_no_outliers[column] = df_adult_cleaned_no_outliers[column].clip(lower=lower_bound, upper=upper_bound)

print("First 5 rows of df_adult_cleaned after outlier capping:")
print(df_adult_cleaned_no_outliers.head())

print("\nSummary statistics of df_adult_cleaned after outlier capping:")
print(df_adult_cleaned_no_outliers.describe())
```

**Reasoning**:
Now that missing values have been imputed and numerical outliers have been addressed, I need to encode the categorical features in `df_adult_cleaned_no_outliers`. I will use one-hot encoding for all object-type columns to convert them into a numerical format suitable for modeling. I will also drop the first category to avoid multicollinearity, create a new DataFrame `df_adult_encoded` for the result, and display its first few rows.



```python
df_adult_encoded = df_adult_cleaned_no_outliers.copy()

# Identify categorical columns for encoding
categorical_cols_adult = df_adult_encoded.select_dtypes(include='object').columns.tolist()

# Apply one-hot encoding to categorical columns, dropping the first category
df_adult_encoded = pd.get_dummies(df_adult_encoded, columns=categorical_cols_adult, drop_first=True)

print("First 5 rows of df_adult after one-hot encoding:")
print(df_adult_encoded.head())

print("Shape of df_adult after one-hot encoding:")
print(df_adult_encoded.shape)
```

**Reasoning**:
Now that categorical features have been one-hot encoded, I will apply Min-Max scaling to all numerical features in the `df_adult_encoded` DataFrame to scale them to a fixed range, typically 0 to 1. I need to import `MinMaxScaler` from `sklearn.preprocessing` for this.



```python
from sklearn.preprocessing import MinMaxScaler

df_adult_minmax_scaled = df_adult_encoded.copy()

# Identify numerical columns for scaling (all except the one-hot encoded 'income_>50K' if it's the target)
numerical_cols_adult = df_adult_minmax_scaled.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Apply Min-Max scaling to the numerical columns
df_adult_minmax_scaled[numerical_cols_adult] = scaler.fit_transform(df_adult_minmax_scaled[numerical_cols_adult])

print("First 5 rows of df_adult after Min-Max scaling:")
print(df_adult_minmax_scaled.head())

print("\nSummary statistics of df_adult after Min-Max scaling:")
print(df_adult_minmax_scaled.describe())
```

**Reasoning**:
Now, I will apply Standard scaling to all numerical features in the encoded DataFrame `df_adult_encoded` to transform the data to have a mean of 0 and a standard deviation of 1. I need to import `StandardScaler` from `sklearn.preprocessing` for this.



```python
from sklearn.preprocessing import StandardScaler

df_adult_standard_scaled = df_adult_encoded.copy()

# Identify numerical columns for scaling (all except the one-hot encoded 'income_>50K' if it's the target)
numerical_cols_adult = df_adult_standard_scaled.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Initialize StandardScaler
s_scaler = StandardScaler()

# Apply Standard scaling to the numerical columns
df_adult_standard_scaled[numerical_cols_adult] = s_scaler.fit_transform(df_adult_standard_scaled[numerical_cols_adult])

print("First 5 rows of df_adult after Standard scaling:")
print(df_adult_standard_scaled.head())

print("\nSummary statistics of df_adult after Standard scaling:")
print(df_adult_standard_scaled.describe())
```

### Summary of Preprocessing Steps for Adult Income Dataset

1.  **Handling Missing Values**:
    - Implicit missing values (represented as '?' or leading/trailing spaces) in categorical columns were identified and cleaned by stripping whitespace.
    - '?' values across all columns were replaced with `np.nan`.
    - Missing values (`np.nan`) in categorical columns (`workclass`, `occupation`, `native-country`) were imputed using the mode of their respective columns.

2.  **Categorical Feature Encoding**:
    - All object-type columns were transformed using one-hot encoding, creating new binary columns for each category. The `drop_first=True` argument was used to avoid multicollinearity.

3.  **Addressing Outliers**:
    - Outliers in all numerical columns (`age`, `fnlwgt`, `education-num`, `capital-gain`, `capital-loss`, `hours-per-week`) were capped using the Interquartile Range (IQR) method. Values below `Q1 - 1.5 * IQR` were set to the lower bound, and values above `Q3 + 1.5 * IQR` were set to the upper bound.

4.  **Min-Max Scaling**:
    - All numerical features in the encoded dataset were scaled to a range between 0 and 1 using `MinMaxScaler`.

5.  **Standard Scaling**:
    - All numerical features in the encoded dataset were also scaled to have a mean of 0 and a standard deviation of 1 using `StandardScaler`.
